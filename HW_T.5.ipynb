```python
df_industry = pd.read_csv(r"C:\Users\gurzh\Downloads\industries(2.0).csv")
```

```python
df['Applied at'] = df['Applied at'].apply(lambda x: x+':00' if len(x.split()[-1].split(':')) == 2 else x)
```

```python
df.drop_duplicates(subset='applicant_id', keep='last', inplace=True) #Task 1
df.shape
```

```python
df['External Rating'].fillna(0, inplace=True)  # Task 2
```

```python
df.fillna({'Education level': 'Середня'}, inplace=True) #Task 3
```

```python
new_df = df.merge(df_industry, on='Industry')
new_df
```

```python
new_df['Applied at'] = pd.to_datetime(new_df['Applied at'], errors='coerce')
new_df

```

```python
new_df['Score1'] = new_df['Score'].where((new_df['Age'] <35) | (new_df['Age'] > 55),  20)  #Condition 1

```

```python
new_df['Day'] = new_df['Applied at'].dt.day_of_week
```

```python
new_df['Score2'] = new_df['Day'].apply(lambda x: 20 if (x!=5) & (x!=6) else 0)  #Condition 2
new_df

```

```python
new_df['Score3'] = new_df['Marital status'].apply(lambda x: 20 if 'Married' in str(x) else 0)   #Condition 3
new_df
```

```python
new_df['Score4'] = new_df['Location'].apply(lambda x: 10 if 'Київ чи область' in str(x) else 0)   #Condition 4
new_df
```

```python
new_df[new_df['Location'] == 'Київ чи область']
```

```python
new_df['Score5']=  np.where(new_df['External Rating'] >=7, 20, np.where(new_df['External Rating'] <=2, -20, 0)) #Condition 5 and 6
new_df
```

```python
new_df
```

```python
new_df['Final_score'] = new_df['Score'] + new_df['Score1'] + new_df['Score2'] + new_df['Score3'] + new_df['Score4'] + new_df['Score5']
new_df
```

```python
new_df['Raiting'] = ((new_df['Final_score']>= 0) &  (new_df['Final_score'] <=100) & (new_df['Amount'] !=0) &(new_df['External Rating'] !=0))*new_df['Final_score']
new_df
```

```python
df1 = new_df[new_df['Raiting'] > 0]
df1
```

```python
df1['Applied at']
```

```python
df1['Week'] = df1['Applied at'].year
```

```python
df1['Week_of_year'] = df1['Applied at'].dt.isocalendar().week.astype(str) + '.' + df1['Applied at'].dt.isocalendar().year.astype(str)
df1
```

```python
df2 = df1.groupby(['Week_of_year'])['Raiting'].mean()
df2
```
